```python
import pandas as pd

# Define the file path and worksheet name
file_path = '/Users/femisokoya/Documents/GitHub/CRA/Companies_Register_Activities_2021-22.xls'
worksheet_name = 'Table_A4'

# Read the Excel file, setting row 5 as the header
df = pd.read_excel(file_path, sheet_name=worksheet_name, header=4)

# Unmerge the merged cells in row 4
for col in df.columns:
    df.at[4, col] = None

# List of column indices to process
column_indices = [3, 7]

# Define rows to update based on your requirements
rows_to_update = [5, 6, 7, 9, 10, 11, 12, 13, 15, 16, 17, 18, 19, 21, 22, 24, 25, 26, 28]

# Iterate over each column index
for i, col_index in enumerate(column_indices):
    # Set the header in row 3 of the current column
    header_label = f'On_reg_{i + 21}' if col_index == 3 else f'Total register_{i + 21}'
    df.iloc[3, col_index] = header_label

    # Loop through the specified rows and set the values in the current column
    for row in rows_to_update:
        # If the column index is 3, set the value to '2021-04-01', else set it to '2022-03-31'
        date_value = '2021-04-01' if col_index == 3 else '2022-03-31'
        df.iat[row, col_index] = date_value

# Make row 5 the column header and delete the top four rows
df = df.iloc[4:].reset_index(drop=True)

# Round all numeric columns to the nearest whole number except 'Effective register'
numeric_columns = df.select_dtypes(include=['number']).columns.difference(['Effective register'])
df[numeric_columns] = df[numeric_columns].round().astype('Int64')

# Rename specific columns using actual positions
df = df.rename(columns={'On_reg_21': 'Number of incorporations'})

# Save the result in '2-cra-result.csv'
df.to_csv('2-cra-result2.csv', index=False)
```
